## Low skills subset

```python
l_df = df2[df2['occupation'] == 'Skill level 1 ~ low']
l_df.head()
```

```python
# Now I want to re-do all the steps I did in the previous work.
# First, I drop the occupation column as I know I am looking only at low occupation workers.

del l_df['occupation']
```

```python
# Again, finding min and max year for each country.

min_max_df = l_df[['year', 'country']].groupby('country').agg(['min', 'max']).reset_index()
min_max_df# Again, finding min and max year for each country.

min_max_df = l_df[['year', 'country']].groupby('country').agg(['min', 'max']).reset_index()
min_max_df
```

```python
# Create a list of countries with only a single data point.

lsame_country_df = min_max_df[min_max_df[('year', 'min')] == min_max_df[('year', 'max')]]
lsame_country_df
```

```python
# Creating 2 DataFrames: one for males and one for females.
l_df2 = l_df[['country', 'sex', 'year', 'monthly_usd']]
l_df2_male = l_df2[l_df2['sex'] == 'Male']
l_df2_female = l_df2[l_df2['sex'] == 'Female']

# See sample data.
l_df2_female.head()
```

```python
# Now joining both male and female dataframes.

l_df2_joint = l_df2_female.merge(l_df2_male, on=['country' ,'year'], suffixes=('_female', '_male'))
l_df2_joint.head()

# Data looks good.
```

```python
# Now I want to calculate the ratio of female to male.

l_df2_joint['ratio'] = l_df2_joint['monthly_usd_female'] / l_df2_joint['monthly_usd_male']
l_df2_joint
```

```python
# Now clean up the Dataframe - remove redundant columns.

del l_df2_joint['sex_female']
del l_df2_joint['sex_male']
```

```python
del l_df2_joint['monthly_usd_female']
del l_df2_joint['monthly_usd_male']

l_df2_joint.head()
```

```python
# Now clean up the min and max table - remove indices.

min_max_df.to_clipboard(index=False)
min_max_df = pd.read_clipboard(sep='\t')
min_max_df = min_max_df.dropna()
min_max_df = min_max_df.rename(columns={'year': 'min_year', 'year.1': 'max_year'})
min_max_df
```

```python
# First joining on the min year.

merged_min = l_df2_joint.merge(min_max_df, left_on=('country', 'year'), right_on=('country','min_year'))
merged_min.head()
```

```python
# Now joining on the max year.
merged_max = l_df2_joint.merge(min_max_df, left_on=('country', 'year'), right_on=('country','max_year'))
merged_max.head()
```

```python
# Now concatenate both DataFrames, remove duplicates and remove min and max year columns as they redundant.

frames = [merged_min, merged_max]
concat_ldf = pd.concat(frames)
del concat_ldf['min_year']
del concat_ldf['max_year']
```

```python
# Dropping duplicates
concat_ldf = concat_ldf.drop_duplicates()
```

```python
concat_ldf.to_csv('low_skill.csv')
```
