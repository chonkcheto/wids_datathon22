```python
# The data doesn't look clean - the gaps between 2014, 2016, 2018 look very big compared to other years.
# Let's look at other countries.

df2t = df2[df2['occupation'] == 'Total']
df2t
```

```python
# Let's see how many unique years I can see per country.

# Make sure I can see all rows.

pd.set_option('display.max_rows', None)

df2t[['year', 'country']].groupby('country').nunique().reset_index()
```

```python
# Let's find the min and max year for each country.

years_df = df2t[['year', 'country']].groupby('country').agg(['min', 'max']).reset_index()

years_df

# There are 136 countries here.
```

```python
# Let's look at full column names.
years_df.columns
```

```python
# First, let's list out the countries where I cannot make a comparison due to a single data point.

same_country_df = years_df[years_df[('year', 'min')] == years_df[('year', 'max')]]
# same_country_df['is_same_year'] == True
same_country_df.shape #.value_counts()

# We have 28 countries with the same year.
```

```python
# For countries without more than 1 year of data
# I want to see which years have the most data.

same_country_df.sort_values(by=('year', 'max'))

# It looks like no particular year stands out.
```

```python
# Find countries with two different max and min years.

diff_country_df = years_df[years_df[('year', 'min')] != years_df[('year', 'max')]]

diff_country_df.shape
```

```python
diff_country_df
```

```python
# Getting rid of the occupation column as it seems redundant.

# Creating 2 DataFrames: one for males and one for females.
df2t2 = df2t[['country', 'sex', 'year', 'monthly_usd']]
df2t2_male = df2t2[df2t2['sex'] == 'Male']
df2t2_female = df2t2[df2t2['sex'] == 'Female']
```

```python
# Joining both DataFrames on country and year.
# This will allow me to see monthly_usd side-by-side and allow me to easily do the ratio calculation.

df2t2_joint = df2t2_female.merge(df2t2_male, on=['country' ,'year'], suffixes=('_female', '_male'))

# Making sure the data looks good.
df2t2_joint.head()
```

```python
# Now I want to calculate the ratio of female to male.

df2t2_joint['ratio'] = df2t2_joint['monthly_usd_female'] / df2t2_joint['monthly_usd_male']
df2t2_joint
```

```python
# Cleaning up the final DataFrame and dropping redundant columns
final_df = df2t2_joint[['country', 'year', 'ratio']]

# Checking it looks good.
final_df.head()
```

```python
final_df.shape
```

```python
# First, I remove multi-level indices before merging.

diff_country_df.to_clipboard(index=False)
diff_country_df=pd.read_clipboard(sep='\t')
diff_country_df = diff_country_df.rename(columns= {'year' : 'min_year',
                        'year.1' : 'max_year'})

diff_country_df = diff_country_df.dropna()
```

```python
diff_country_df.head()
```

```python
# Checking data types of the columns.
diff_country_df.dtypes
```

```python
# Looks like years got converted into strings. I will now convert them into integers.

diff_country_df['min_year'] = diff_country_df['min_year'].astype(int)
diff_country_df['max_year'] = diff_country_df['max_year'].astype(int)

diff_country_df.dtypes
```

```python
# Check for the data types in the final_df

final_df.dtypes

# Seems correct, no conversion needed.
```

```python
# First, I join with the ratio DataFrame on min year.

min_year_final = final_df.merge(diff_country_df, left_on=('country', 'year'), right_on=('country', 'min_year'))
min_year_final.head()
```

```python
# Do the same for the max year.

max_year_final = final_df.merge(diff_country_df, left_on=('country', 'year'), right_on=('country', 'max_year'))
max_year_final.head()
```

```python
# Now I can do a concatenation on these two.

frames = [min_year_final, max_year_final]
concat_df = pd.concat(frames)

# Next steps: reset index to be able to order by country and year.
concat_df = concat_df.reset_index()
del concat_df['index']
```

```python
final_total_df = concat_df.sort_values(by=(['country', 'year']))

del final_total_df['min_year']
del final_total_df['max_year']
```

```python
# Above I produced a DataFrame with data for countries that have two different years available.
# Now I want to add data for countries that only have 1 year of data available.

final_total_df
```

```python
# Converting the countries into a list (countries with one year of data available only).
same_country_list = (same_country_df['country']).tolist()
same_country_list
```

```python
# Getting a subset of countries that only have one year.

final_oneyear = final_df.loc[final_df.apply(lambda x: x.country in same_country_list, axis=1)]
frames = [final_oneyear, final_total_df]
super_final_df = pd.concat(frames)
super_final_df = super_final_df.sort_values(by='country')
super_final_df.to_csv("total_df.csv")
```
